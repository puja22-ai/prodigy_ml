```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
image_filename = list(uploaded.keys())[0]

# Read the image using OpenCV
image = cv2.imread(image_filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# Split into R, G, B channels
R, G, B = cv2.split(image)

# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Display the images
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

axes[0].imshow(R, cmap="Reds")
axes[0].set_title("Red Channel")
axes[0].axis("off")

axes[1].imshow(G, cmap="Greens")
axes[1].set_title("Green Channel")
axes[1].axis("off")

axes[2].imshow(B, cmap="Blues")
axes[2].set_title("Blue Channel")
axes[2].axis("off")

axes[3].imshow(gray_image, cmap="gray")
axes[3].set_title("Grayscale Image")
axes[3].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
image_filename = list(uploaded.keys())[0]

# Read the image
image = cv2.imread(image_filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Apply Canny Edge Detection with different thresholds
edges_low = cv2.Canny(gray_image, 50, 100)   # Lower thresholds (detects more edges)
edges_med = cv2.Canny(gray_image, 100, 200)  # Medium thresholds (balanced)
edges_high = cv2.Canny(gray_image, 150, 250) # Higher thresholds (detects fewer edges)

# Display the results
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(edges_low, cmap="gray")
axes[1].set_title("Canny Edges (Low Thresholds: 50, 100)")
axes[1].axis("off")

axes[2].imshow(edges_med, cmap="gray")
axes[2].set_title("Canny Edges (Medium Thresholds: 100, 200)")
axes[2].axis("off")

axes[3].imshow(edges_high, cmap="gray")
axes[3].set_title("Canny Edges (High Thresholds: 150, 250)")
axes[3].axis("off")

plt.show()
```

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# 📌 Step 1: Upload and Load Image
print("Upload your image...")
uploaded = files.upload()
image_filename = list(uploaded.keys())[0]
image = cv2.imread(image_filename)

if image is None:
    print("⚠️ Error: Image not found! Check filename and re-upload.")
else:
    print("✅ Image successfully loaded!")

# Convert image to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 📌 Step 2: Download YOLOv3 Weights and Config (If Not Already Present)
import os
yolo_dir = "yolo"
if not os.path.exists(yolo_dir):
    os.makedirs(yolo_dir)

# YOLOv3 Configuration & Weights
yolo_cfg_url = "https://github.com/pjreddie/darknet/blob/master/cfg/yolov3.cfg?raw=true"
yolo_weights_url = "https://pjreddie.com/media/files/yolov3.weights"
yolo_names_url = "https://raw.githubusercontent.com/pjreddie/darknet/master/data/coco.names"

cfg_path = os.path.join(yolo_dir, "yolov3.cfg")
weights_path = os.path.join(yolo_dir, "yolov3.weights")
names_path = os.path.join(yolo_dir, "coco.names")

# Download files if they don’t exist
if not os.path.exists(cfg_path):
    !wget -O {cfg_path} {yolo_cfg_url}
if not os.path.exists(weights_path):
    !wget -O {weights_path} {yolo_weights_url}
if not os.path.exists(names_path):
    !wget -O {names_path} {yolo_names_url}

# 📌 Step 3: Load YOLO Model
net = cv2.dnn.readNet(weights_path, cfg_path)
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

# Load class labels
with open(names_path, "r") as f:
    classes = f.read().strip().split("\n")

# 📌 Step 4: Prepare Image for YOLO
height, width = image.shape[:2]
blob = cv2.dnn.blobFromImage(image, 1/255.0, (416, 416), swapRB=True, crop=False)
net.setInput(blob)

# 📌 Step 5: Perform Object Detection
detections = net.forward(output_layers)

# 📌 Step 6: Process YOLO Detections
conf_threshold = 0.5  # Confidence threshold
nms_threshold = 0.4   # Non-maximum suppression threshold

boxes = []
confidences = []
class_ids = []

for output in detections:
    for detection in output:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]

        if confidence > conf_threshold:
            center_x, center_y, w, h = (detection[:4] * np.array([width, height, width, height])).astype("int")
            x = int(center_x - (w / 2))
            y = int(center_y - (h / 2))

            boxes.append([x, y, w, h])
            confidences.append(float(confidence))
            class_ids.append(class_id)

# Apply Non-Maximum Suppression
indices = cv2.dnn.NMSBoxes(boxes, confidences, conf_threshold, nms_threshold)


# Convert image to RGB for display
detected_image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 📌 Step 8: Show Result
plt.figure(figsize=(10, 6))
plt.imshow(detected_image_rgb)
plt.axis("off")
plt.title("YOLO Object Detection")
plt.show()
```
