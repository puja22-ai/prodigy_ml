```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from google.colab import files
import cv2
import os

# 📌 Upload an image file
uploaded = files.upload()

# Get file path
image_path = list(uploaded.keys())[0]

# Load a pre-trained deep learning model (MobileNetV2)
model = tf.keras.applications.MobileNetV2(weights='imagenet')

# Load and preprocess the image
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = tf.keras.applications.mobilenet_v2.preprocess_input(img_array)
    return img, img_array

# Classify the image
def classify_food(image_path):
    img, img_array = preprocess_image(image_path)
    predictions = model.predict(img_array)
    decoded_predictions = tf.keras.applications.mobilenet_v2.decode_predictions(predictions, top=3)[0]
    return img, decoded_predictions

# Define a simple calorie estimation lookup table
calorie_lookup = {
    "cheeseburger": 303,
    "hotdog": 290,
    "pizza": 285,
    "spaghetti": 221,
    "salad": 150,
    "sushi": 200,
    "steak": 679,
    "apple": 95,
    "banana": 105,
    "carrot": 25,
    "broccoli": 55
}

# Process the uploaded image
img, predictions = classify_food(image_path)

# Get the top prediction
top_prediction = predictions[0][1].replace("_", " ")  # Format name
confidence = predictions[0][2] * 100  # Convert to percentage
calories = calorie_lookup.get(top_prediction.lower(), "Unknown")  # Get estimated calories

# Display results
plt.imshow(img)
plt.axis("off")
plt.title(f"Prediction: {top_prediction} ({confidence:.2f}%)\nEstimated Calories: {calories} kcal")
plt.show()

# Print detailed predictions
for pred in predictions:
    print(f"{pred[1]}: {pred[2]*100:.2f}% confidence")
```
